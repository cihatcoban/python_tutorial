```python
import requests
from bs4 import BeautifulSoup
import wikipedia
import urllib3
header={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}
url="https://en.wikipedia.org/wiki/Metropolis_of_Ephesus#:~:text=Until%20the%204th%20century%20AD,in%20the%20course%20of%20time."
get=requests.get(url,headers=header)
content=get.content
soup=BeautifulSoup(content,"html.parser")
titles=soup.find_all("div",{"class":"mw-parser-output"})
for i in titles:
    i = i.text
    print(i)
print(get.status_code)
```
