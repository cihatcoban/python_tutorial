```python
print
("""
*********************HESAP MAKİNESİ***********************

------>İŞLEMLER NUMARASI<-------

Toplama = 1
Çıkarma = 2
Çarpma = 3
Bölme = 4
Üs almak için = 5
Sayının karekökünü almak için = 6 
Sayının logaritması için = 7
dereceyi radyana çevirmek için = 8
radyanı dereceye çevirmek için = 9
sinüs ü bulmak için = 10
cos u bulmak için = 11
tanjantı bulmak için = 12
cotanjantı bulmak için = 13
Çıkmak için q ya basın...

""")
import math
import time
while True:
    islem=int(input("lütfen bir işlem numarası : "))
    #çıkış işlemi
    if (islem=="q"):
        print("çıkış yapılıyor....")
        time.sleep(1)
        print("çıkış yapıldı")
        break
    # toplama işlemi
    elif (islem==1):
        x=float(input("1.sayı: "))
        y=float(input("2.sayı: "))
        print("sonuç hesaplanıyor...")
        time.sleep(1)
        print("{} + {} = {}".format(x,y,x+y))
    # çıkarma işlemi
    elif (islem==2):
        x=float(input("1.sayı: "))
        y=float(input("2.sayı: "))
        print("sonuç hesaplanıyor...")
        time.sleep(1)
        print("{} - {} = {}".format(x,y,x-y))
    #çarpma işlemi
    elif (islem==3):
        x=float(input("1.sayı: "))
        y=float(input("2.sayı: "))
        print("sonuç hesaplanıyor...")
        time.sleep(1)
        print("{} * {} = {}".format(x,y,x*y))
    # bölme işlemi
    elif (islem==4):
        x=float(input("1.sayı: "))
        y=float(input("2.sayı: "))
        print("sonuç hesaplanıyor...")
        time.sleep(1)
        print("{} / {} = {}".format(x,y,x/y))
    # üs alma işlemi
    elif (islem==5):
        x=float(input("taban: "))
        y=float(input("üs: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.pow(x,y)
        print("{} üssü {} = {}".format(x,y,z))
    # kök alma işlemi
    elif (islem==6):
        x=float(input("sayı: "))
        print("hesaplanıyor...")
        time.sleep(1)
        y=math.sqrt(x)
        print("{} kökü = {}".format(x,y))
    # logaritma alma
    elif (islem==7):
        x=float(input("sayı: "))
        y=float(input("taban: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.log(x,y)
        print("{} sayısının {} tabanında = {}".format(x,y,z))
    # dereceyi radyana çevirme işlemi
    elif (islem==8):
        x=float(input("derece: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.degrees(x)
        print("{} derece = {} radyan".format(x,z))
    # radyanı dereceye çevirme işlemi
    elif (islem==9):
        x=float(input("radyan: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.radians(x)
        print("{} ^ {} = {}".format(x,z))
    # sin bulma işlemi
    elif (islem==10):
        x=float(input("sin: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.sin(x)
        print("sin {} = {}".format(x,z))
    # cos bulma işlemi
    elif (islem==11):
        x=float(input("cos: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.cos(x)
        print("cos {}  = {}".format(x,z))
    # tan bulma işlemi
    elif (islem==12):
        x=float(input("tan: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.tan(x)
        print("tan {} = {}".format(x,z))
    # cot bulma işlemi
    elif (islem==13):
        x=float(input("cot: "))
        print("hesaplanıyor...")
        time.sleep(1)
        z=math.cos(x)/math.sin
        print("cot {} = {}".format(x,z))
```
